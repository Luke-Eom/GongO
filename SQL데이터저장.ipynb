```python
import pandas as pd

# filename = 'All_Assembled PC.xlsx'
file_name = ["Assembled_PC", "Part_Case", "Part_Cooler", "Part_CPU", "Part_GPU", "Part_HDD", "Part_MainBoard", "Part_Monitor", "Part_Power", "Part_RAM", "Part_SSD"]

for i in file_name:
    globals()[i] = pd.read_excel(i+".xlsx", engine = 'openpyxl', index_col=0)
    globals()[i] = globals()[i].astype({'제품 코드':'str'})
    globals()[i]['제품명'] = globals()[i]['제품명'] + '/' + globals()[i]['제품 코드']
    globals()[i]['최저가'] = globals()[i]['최저가'] + '원'
    globals()[i].drop_duplicates(inplace=True)
    globals()[i]['카테고리'] = i.split('_')[1]
```

```python
df = pd.concat([globals()['Assembled_PC'], globals()['Part_Case'], globals()['Part_Cooler'], globals()['Part_CPU'], globals()['Part_GPU'], globals()['Part_HDD'], globals()['Part_MainBoard'], globals()['Part_Monitor'], globals()['Part_Power'], globals()['Part_RAM'], globals()['Part_SSD']])
product_info = df[['제품명', '제조회사', '카테고리']]
product = df[['제품이미지URL', '가격비교 링크', '최저가', '제품명']]
```
