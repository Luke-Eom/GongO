```python
import pandas as pd

filename = 'All_Assembled PC.xlsx'

df = pd.read_excel(filename, engine = 'openpyxl', index_col=0)

df.head()
```

```python
pd.set_option('display.max_columns', 100)

df.head(3)
```

```python
df = df.astype({'제품 코드' : 'str'})
df.dtypes
```

```python
df['제품명'] = df['제품명'] + ' / ' + df['제품 코드']
# df.drop(['제품 코드'], axis=1, inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index' : 'id'}, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
mv = df.values.tolist()
l = []
for idx, i in enumerate(mv):
    tmpstr = ""
    if i[12] == i[12] :
        tmpstr += '메모리 / ' + str(i[12])
    if i[13] == i[13] :
        tmpstr += ' / ' + str(i[13])
    if i[14] == i[14] :
        tmpstr += ' / ' + str(i[14])
    if i[16] == i[16] :
        tmpstr += ' / HDD / ' + str(i[16])
    l.append(tmpstr)
df['상세 정보'] = l
df['카테고리'] = '완제품'
df
```
